```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate sample dataset of length 90
dates = pd.date_range(start="2024-01-01", periods=90, freq="D")
sales = np.random.randint(200, 1500, size=90)

# Repeat categories so total = 90
categories = (["Electronics", "Clothing", "Groceries"] * 30)[:90]

# Create DataFrame
df = pd.DataFrame({
    "Date": dates,
    "Sales": sales,
    "Category": categories
})

df['Month'] = df['Date'].dt.to_period('M')
df['Quarter'] = df['Date'].dt.quarter

df.head()
```
