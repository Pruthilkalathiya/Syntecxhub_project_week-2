```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate sample dataset of 90 days
dates = pd.date_range(start="2024-01-01", periods=90, freq="D")
sales = np.random.randint(200, 1500, size=90)

# Repeat categories to make total = 90 rows
categories = (["Electronics", "Clothing", "Groceries"] * 30)[:90]

# Create DataFrame
df = pd.DataFrame({
    "Date": dates,
    "Sales": sales,
    "Category": categories
})

# Create additional columns
df['Month'] = df['Date'].dt.to_period('M')
df['Quarter'] = df['Date'].dt.quarter
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram
plt.figure(figsize=(8,5))
sns.histplot(df["Sales"], kde=False)
plt.title("Sales Distribution - Histogram")
plt.xlabel("Sales")
plt.tight_layout()
plt.savefig("sales_histogram.png")
plt.show()

# KDE Plot
plt.figure(figsize=(8,5))
sns.kdeplot(df["Sales"], shade=True)
plt.title("Sales Distribution - KDE")
plt.xlabel("Sales")
plt.tight_layout()
plt.savefig("sales_kde.png")
plt.show()

# Boxplot (Outlier detection)
plt.figure(figsize=(8,5))
sns.boxplot(df["Sales"])
plt.title("Sales Boxplot (Outliers)")
plt.tight_layout()
plt.savefig("sales_boxplot.png")
plt.show()

# Compare distributions by quarter
plt.figure(figsize=(9,6))
sns.boxplot(x=df["Quarter"], y=df["Sales"])
plt.title("Sales Comparison Across Quarters")
plt.tight_layout()
plt.savefig("quarter_comparison.png")
plt.show()
```
